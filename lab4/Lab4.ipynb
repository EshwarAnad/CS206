```python
import random

import time 
from numpy.random import seed 
from numpy.random import randint 
import matplotlib.pyplot as plt 

def rand(start,end,num):
    x=[]
    for j in range(num): 
        x.append(random.randint(start, end)) 
  
    return x
  
start=1
end=10000
num=10
#print(rand(start,end,num)) 
y = rand(start,end,num)
z = rand(start,end,num)
print('y :',y)
print('z :',z)



# x = list(map(int, input("Enter Multi Array: ").split())) 
# print('numbers',x)

def insertionSort(y):
    for i in range(1,len(y)):
        key = y[i]
        j = i-1
        while j >=0 and key < y[j] : 
                y[j+1] = y[j] 
                j -= 1
                y[j+1] = key

                
def partition(z,low,high): 
    i = ( low-1 )         
    pivot = z[high]     
  
    for j in range(low , high): 
  
        if   z[j] <= pivot: 
          
            i = i+1 
            z[i],z[j] = z[j],z[i] 
  
    z[i+1],z[high] = z[high],z[i+1] 
    return ( i+1 ) 

  
def quickSort(z,low,high): 
    if low < high: 
  
 
        pi = partition(z,low,high) 
  
        quickSort(z, low, pi-1) 
        quickSort(z, pi+1, high)
        
        
n = len(y) 
#insertionSort(y)
#print('Insertion Sort of y :',y)
#quickSort(z,0,n-1) 
#print ("Quick Sort of z is:",z) 




elements = y
times = y
  
start = time.clock() 
insertionSort(y) 
print('Insertion Sort of y :',y)
end = time.clock() 
  
elements.append(len(y)) 
print(len(y), "Elements Sorted by Insertion Sort in ", end-start) 

times.append(end-start) 
  
plt.xlabel('List Length') 
plt.ylabel('Time Complexity') 
plt.plot(elements, times, label ='Insertion Sort') 
plt.grid() 
plt.legend() 
plt.show()



elements = z
times = z 
  
start = time.clock() 
quickSort(z,0,n-1) 
print ("Quick Sort of z is:",z) 
end = time.clock() 
  
elements.append(len(z)) 
print(len(z), "Elements Sorted by QuickSort in ", end-start) 

times.append(end-start) 
  
plt.xlabel('List Length') 
plt.ylabel('Time Complexity') 
plt.plot(elements, times, label ='Quick Sort') 
plt.grid() 
plt.legend() 
plt.show() 
```
